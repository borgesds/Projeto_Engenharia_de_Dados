# Extração dos dados Enade

---------------------------------------------------------------------------------------------

```python
# importar bibliotecas
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# criar um diretorio para os dados
os.makedirs('./enade2021', exist_ok=True)
```

```python
# define a url
url = 'https://download.inep.gov.br/microdados/microdados_enade_2021_LGPD.zip'

# download 
filebytes = BytesIO(
    requests.get(url).content
)

# extrair conteudo
myzip = zipfile.ZipFile(filebytes)
myzip.extractall('./enade2021')
```

## Verificando os dados baixados

```python
enade = pd.read_csv(
    './enade2021/microdados_enade_2021/microdados_Enade_2021_LGPD/2.DADOS/microdados2021_arq4.txt',
    sep=';', decimal=','
)
```
